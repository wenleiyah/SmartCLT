create dataframe for ALGO

```python
import pandas as pd

# Create a dictionary with column names as keys and empty lists as values
data = {
    'NO': [],
    'rotation_det': [],
    'log_left': [],
    'log_right': []
}

# Populate the data dictionary from the solution list
for i in range(len(solution[0])):
    data['NO'].append(solution[0][i])
    data['rotation_det'].append(solution[1][i])

    left_index = (i - 1) % len(solution[0])  # Wrap around to the end if i-1 goes out of range
    data['log_left'].append(solution[0][left_index])

    right_index = (i + 1) % len(solution[0])  # Wrap around to the beginning if i+1 goes out of range
    data['log_right'].append(solution[0][right_index])

# Create a DataFrame from the data dictionary
df = pd.DataFrame(data)

df
```

```python
# Save the DataFrame as a CSV file in the /content/ directory
output_path = '/content/data.csv'
df.to_csv(output_path, index=False)
```
